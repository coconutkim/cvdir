```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('../../img/girl.jpg')

plt.imshow(img[:,:,::-1])       # 이미지 컬러 채널 변경해서 표시 --- ①
# 모든 행과 열을 선택하고 마지막에 색상 채널을 역순으로 뒤집는다
plt.xticks([])                  # x좌표 눈금 제거         ---②     
plt.yticks([])                  # y좌표 눈금 제거         ---③

# plt.imshow(img) # 이미지 표시
plt.show()
```

# 왜 변환 과정이 필요한가

### 색상 채널 순서의 기술적 배경

1. **OpenCV (BGR)**:
   - **기술적 배경**: OpenCV는 C++로 처음 개발되었고, 초기 설계에서는 색상 데이터를 BGR 순서로 저장했습니다. 이는 다양한 하드웨어와 호환성 문제를 다루기 위해 선택된 방식입니다. 특히, 비디오 카메라와 같은 하드웨어 장치에서 이 순서를 사용하는 경우가 많았기 때문에 OpenCV가 BGR을 채택했습니다.
   - **OpenCV의 BGR 사용**: OpenCV의 많은 기능이 비디오 처리와 관련이 있으며, 오래된 비디오 표준에서는 BGR을 사용하기도 했습니다. 이런 역사적 이유로 BGR 순서가 사용되었습니다.

2. **Matplotlib (RGB)**:
   - **기술적 배경**: Matplotlib과 다른 많은 데이터 시각화 도구들은 RGB 순서를 채택하고 있습니다. 이는 컴퓨터 그래픽스와 디지털 이미징에서 RGB가 표준으로 널리 사용되기 때문입니다. RGB는 웹에서 사용되는 이미지 표준이며, 많은 이미지 파일 포맷(예: JPEG, PNG)도 RGB를 기본 색상 순서로 사용합니다.
   - **Matplotlib의 RGB 사용**: Matplotlib은 과학적 시각화와 데이터 분석 도구로 설계되었기 때문에, RGB 순서가 더 자연스러운 선택이었습니다. RGB는 디스플레이 장치의 기본 색상 모델과도 잘 맞아떨어집니다.

### 색상 채널 순서의 변환

이러한 차이는 색상 채널 순서가 서로 다르기 때문에 발생합니다. 각 라이브러리가 채택한 색상 채널 순서는 다음과 같은 이유로 결정되었습니다:

- **호환성**: OpenCV는 다양한 하드웨어와 프로토콜을 지원하기 위해 BGR을 사용했습니다.
- **표준**: Matplotlib과 많은 데이터 시각화 도구는 RGB를 표준으로 사용합니다.

### 변환 방법

이러한 색상 순서의 차이를 해결하기 위해, 이미지를 OpenCV에서 읽고 Matplotlib으로 표시할 때는 색상 채널 순서를 변환해야 합니다. NumPy의 슬라이싱 기능을 이용하여 BGR을 RGB로 바꾸는 것이죠.

### 요약

색상 채널 순서의 차이는 각 라이브러리의 역사적 배경과 기술적 설계 선택에 기인합니다. OpenCV는 BGR을, Matplotlib은 RGB를 사용하며, 이 차이를 처리하기 위해 색상 채널 순서를 변환하는 과정이 필요합니다. 이러한 차이는 실제 오류가 아니라 각 도구의 설계와 표준에 따른 결과입니다.

# 비디오와 이미지의 색상 채널 순서

1. **비디오 (BGR)**:
   - **OpenCV**에서는 비디오 프레임을 처리할 때 기본적으로 BGR 순서를 사용합니다. 이는 비디오 캡처 장치와의 호환성을 유지하기 위해 선택된 방식입니다.
   - 비디오 데이터를 다룰 때는 BGR 순서로 색상 채널이 저장되며, 이는 하드웨어와의 효율적인 호환성 문제와 관련이 있습니다.

2. **정지 이미지 (RGB 또는 BGR)**:
   - **정지 이미지**는 보통 RGB 순서를 사용하는 경우가 많습니다. 이는 디지털 이미지 파일 포맷(예: JPEG, PNG)이 RGB 색상 모델을 기본으로 사용하기 때문입니다.
   - OpenCV에서는 이미지도 BGR 순서로 읽어들이지만, 일부 이미지 포맷을 다룰 때는 RGB를 사용하기도 합니다. 

### 비디오와 이미지 처리에서 색상 채널 순서의 차이

- **비디오 캡처**: OpenCV에서 비디오를 처리할 때, 색상 채널 순서는 BGR입니다. 비디오 프레임을 추출하거나 표시할 때는 이 순서에 맞추어야 합니다.
- **정지 이미지**: 이미지를 파일에서 읽어오거나 처리할 때는 종종 RGB 순서를 사용합니다. 그러나 OpenCV에서는 BGR로 읽어들이기 때문에, Matplotlib 같은 시각화 도구에서 이미지를 올바르게 표시하려면 RGB로 변환해야 합니다.
